**Why the Memory Addresses are Different:**
- **Temporary Instance (id(Cheff())):**
> Each call to Cheff() creates a new, unique instance of the Cheff class.
When you call id(Cheff()), a new instance is created temporarily for the id function call, and this instance is not stored or referenced after the id call.
Since it is a new instance, it has a unique memory address each time you call Cheff().

- **Persistent Instance (id(user_1)):**
> When you call user_1 = Cheff(), an instance of Cheff is created and stored in the variable user_1.
The id(user_1) call retrieves the memory address of this specific, persistent instance.

```python
class Car:
    # 1. мғқм„ұмһҗ: к°қмІҙк°Җ л§Ңл“Өм–ҙм§Ҳ л•Ң нҳём¶ңлҗҳлҠ” н•ЁмҲҳ
    def __init__(self, brand, model, color):
        self.brand = brand  # лёҢлһңл“ң
        self.model = model  # лӘЁлҚёлӘ…
        self.color = color  # мғүмғҒ
        self.speed = 0      # нҳ„мһ¬ мҶҚлҸ„ (кё°ліёк°’ 0)

    # 2. л©”м„ңл“ң: мһҗлҸҷм°Ёмқҳ лҸҷмһ‘ (к°ҖмҶҚ)
    def accelerate(self, amount):
        self.speed += amount
        print(f"{self.model}мқҙ(к°Җ) {amount}km/h к°ҖмҶҚн•ҳм—¬ нҳ„мһ¬ {self.speed}km/hмһ…лӢҲлӢӨ.")

    # 3. л©”м„ңл“ң: мһҗлҸҷм°Ёмқҳ лҸҷмһ‘ (м •м§Җ)
    def stop(self):
        self.speed = 0
        print(f"{self.model}мқҙ(к°Җ) л©Ҳм·„мҠөлӢҲлӢӨ.")
```

```python
# 'my_car'лқјлҠ” мӢӨм ң мһҗлҸҷм°Ё к°қмІҙ мғқм„ұ
my_car = Car("Hyundai", "Ionic 6", "White")

# 'your_car'лқјлҠ” лҳҗ лӢӨлҘё мһҗлҸҷм°Ё к°қмІҙ мғқм„ұ
your_car = Car("Tesla", "Model 3", "Black")

# к°Ғ мһҗлҸҷм°Ёмқҳ м •ліҙ м¶ңл Ҙ
print(f"лӮҙ м°ЁлҠ” {my_car.brand} {my_car.model}мһ…лӢҲлӢӨ.")
# м¶ңл Ҙ: лӮҙ м°ЁлҠ” Hyundai Ionic 6мһ…лӢҲлӢӨ.

# мһҗлҸҷм°Ё лҸҷмһ‘мӢңнӮӨкё°
my_car.accelerate(50)  # Ionic 6мқҙ 50km/hлЎң лӢ¬лҰҪлӢҲлӢӨ.
your_car.accelerate(80) # Model 3лҠ” 80km/hлЎң лӢ¬лҰҪлӢҲлӢӨ.
```

```python
from faker import Faker

class User:
    def __init__(self, user_id, user_name, user_age):
        self.id = user_id
        self.name = user_name
        self.age = user_age
        self.following = []  # List of users this user is following
        self.followers = []  # List of users following this user

    def follow(self, target_user):
        if target_user.name not in self.following:
            self.following.append(target_user.name)
            target_user.followers.append(self.name)
            print(f"{self.name} is now following {target_user.name}.")
        else:
            print(f"{self.name} is already following {target_user.name}.")

fake = Faker()
users_1000 = []

for i in range(1, 1001):
    user_id = str(i).zfill(4) # Ensure unique and formatted IDs
    user_name = fake.name()
    user_age = fake.random_int(min=18, max=90)

    user = User(user_id, user_name, user_age)
    users_1000.append(user)

print(f"Generated {len(users_1000)} users.")

# Display some sample data
print("\nFirst 5 generated users:")
for i in range(5):
    print(f"ID: {users_1000[i].id}, Name: {users_1000[i].name}, Age: {users_1000[i].age}")

print("\nLast 5 generated users:")
for i in range(995, 1000):
    print(f"ID: {users_1000[i].id}, Name: {users_1000[i].name}, Age: {users_1000[i].age}")
```

## **кіјм ң 2. Tomмқҙ JerryлҘј м«’м•„к°ҖлҠ” н–үмң„лҘј method(н•ЁмҲҳ)лЎң л§Ңл“Өм–ҙ ліҙмһҗ(10лІҲ following)**

### **class User:**
> **def __init __ (self):**
>   - ##### Initialize attributes
>   - **HAS (Attiributes)**

> **def following_mode():**
>   - ##### change attributes (following, follower) for chasing
>   - **Does (Method, Change the attributes)**

<img src= 'https://pbs.twimg.com/profile_images/810136393/tom-jerry-pic_400x400.jpg' width=300>

```python
class User:
    def __init__(self, name):
        self.name = name
        self.followers = 0
        self.following = 0

    def follow(self, target_user):
        self.following += 1
        target_user.followers += 1
        print(f"вҖ» {self.name}мқҙ(к°Җ) {target_user.name}мқ„(лҘј) м«“м•„к°‘лӢҲлӢӨ!\n(нҳ„мһ¬ {self.name}мқҳ м¶”кІ© нҡҹмҲҳ: {self.following}нҡҢ)")

tom = User("Tom")
jerry = User("Jerry")

print("м¶”кІ©м „ мӢңмһ‘!\n")

for i in range(10):
    tom.follow(jerry)

print("\n--- мөңмў… кІ°кіј ---")
print(f"{tom.name} нҢ”лЎңмһү(м¶”кІ© мӨ‘): {tom.following}нҡҢ")
print(f"{jerry.name} нҢ”лЎңмӣҢ(лҸ„л§қ мӨ‘): {jerry.followers}нҡҢ")
```
