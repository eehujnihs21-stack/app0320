```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 1. 설정 (네이버가 봇으로 인식 못하게 User-Agent 강화)
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36")

driver = webdriver.Chrome(options=options)

try:
    # 2. 신작 페이지 접속
    url = "https://comic.naver.com/webtoon?tab=new"
    driver.get(url)

    # 3. 데이터 로딩을 위해 강제로 7초 대기 (0개 뜨는 거 방지)
    print("🌐 네이버 서버와 밀당 중... (7초만 기다려주세요)")
    time.sleep(7)

    # 4. 'ContentList'가 포함된 모든 li 요소를 더 넓은 범위에서 검색
    # 클래스명 변동을 대비해 별표(*) 연산자 사용
    items = driver.find_elements(By.CSS_SELECTOR, "li[class*='ContentList__content_list_item']")

    if not items:
        # 만약 그래도 0개라면, 다른 선택자로 재시도
        items = driver.find_elements(By.CSS_SELECTOR, ".component_wrap li")

    print(f"✅ 총 {len(items)}개의 웹툰 데이터를 확보했습니다!\n")

    # 5. 데이터 추출 및 상세페이지 이동
    for i in range(min(5, len(items))):
        try:
            # 제목과 작가는 각 li 안의 span이나 strong에 들어있음
            title = items[i].find_element(By.CSS_SELECTOR, "[class*='title']").text
            author = items[i].find_element(By.CSS_SELECTOR, "[class*='author']").text
            link = items[i].find_element(By.TAG_NAME, "a").get_attribute("href")

            # 상세 페이지 진입
            driver.execute_script(f"window.open('{link}');")
            driver.switch_to.window(driver.window_handles[-1])
            time.sleep(3) # 상세페이지 로딩 대기

            try:
                # 줄거리 추출 (summary 글자가 들어간 모든 요소 중 첫 번째)
                summary = driver.find_element(By.CSS_SELECTOR, "[class*='summary']").text
            except:
                summary = "줄거리를 찾을 수 없습니다."

            print(f"📌 제목: {title}")
            print(f"👤 작가: {author}")
            print(f"📖 내용: {summary}")
            print("-" * 50)

            driver.close()
            driver.switch_to.window(driver.window_handles[0])

        except Exception as e:
            continue

except Exception as e:
    print(f"❌ 에러 발생: {e}")
finally:
    driver.quit()
```
