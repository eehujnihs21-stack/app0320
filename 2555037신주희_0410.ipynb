```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 1. 크롬 옵션 (최대한 안정적으로 설정)
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

# 드라이버 실행
driver = webdriver.Chrome(options=options)

try:
    # 2. 신작 페이지 접속
    url = "https://comic.naver.com/webtoon?tab=new"
    driver.get(url)

    # 3. 데이터가 뜰 때까지 최대 10초 대기 (이게 핵심!)
    wait = WebDriverWait(driver, 10)
    # 신작 리스트 아이템이 나타날 때까지 기다림
    items = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "li[class*='ContentList__content_list_item']")))

    print(f"✅ 신작 웹툰 {len(items)}개를 찾았습니다. 내용을 추출합니다...\n")

    webtoon_data = []

    # 4. 리스트에서 제목, 작가, 링크 추출
    for item in items[:10]: # 너무 많으면 오래 걸리니 상위 10개만
        try:
            # 클래스명 뒤의 난수를 무시하기 위해 'contains' 기법 사용
            title_el = item.find_element(By.CSS_SELECTOR, "[class*='ContentTitle__title']")
            author_el = item.find_element(By.CSS_SELECTOR, "[class*='ContentAuthor__author']")
            link_el = item.find_element(By.TAG_NAME, "a")

            webtoon_data.append({
                "title": title_el.text,
                "author": author_el.text,
                "link": link_el.get_attribute("href")
            })
        except:
            continue

    # 5. 상세 페이지 들어가서 줄거리(내용) 가져오기
    for data in webtoon_data:
        driver.get(data['link'])
        # 줄거리가 나타날 때까지 대기
        summary_el = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "[class*='EpisodeListInfo__summary']")))

        print(f"📌 제목: {data['title']}")
        print(f"👤 작가: {data['author']}")
        print(f"📖 내용: {summary_el.text}")
        print("-" * 60)

except Exception as e:
    print(f"❌ 에러 발생: {e}")

finally:
    driver.quit()
    print("\n✅ 작업 완료!")
```

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 1. 설정 (네이버가 봇으로 인식 못하게 User-Agent 강화)
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36")

driver = webdriver.Chrome(options=options)

try:
    # 2. 신작 페이지 접속
    url = "https://comic.naver.com/webtoon?tab=new"
    driver.get(url)

    # 3. 데이터 로딩을 위해 강제로 7초 대기 (0개 뜨는 거 방지)
    print("🌐 네이버 서버와 밀당 중... (7초만 기다려주세요)")
    time.sleep(7)

    # 4. 'ContentList'가 포함된 모든 li 요소를 더 넓은 범위에서 검색
    # 클래스명 변동을 대비해 별표(*) 연산자 사용
    items = driver.find_elements(By.CSS_SELECTOR, "li[class*='ContentList__content_list_item']")

    if not items:
        # 만약 그래도 0개라면, 다른 선택자로 재시도
        items = driver.find_elements(By.CSS_SELECTOR, ".component_wrap li")

    print(f"✅ 총 {len(items)}개의 웹툰 데이터를 확보했습니다!\n")

    # 5. 데이터 추출 및 상세페이지 이동
    for i in range(min(5, len(items))):
        try:
            # 제목과 작가는 각 li 안의 span이나 strong에 들어있음
            title = items[i].find_element(By.CSS_SELECTOR, "[class*='title']").text
            author = items[i].find_element(By.CSS_SELECTOR, "[class*='author']").text
            link = items[i].find_element(By.TAG_NAME, "a").get_attribute("href")

            # 상세 페이지 진입
            driver.execute_script(f"window.open('{link}');")
            driver.switch_to.window(driver.window_handles[-1])
            time.sleep(3) # 상세페이지 로딩 대기

            try:
                # 줄거리 추출 (summary 글자가 들어간 모든 요소 중 첫 번째)
                summary = driver.find_element(By.CSS_SELECTOR, "[class*='summary']").text
            except:
                summary = "줄거리를 찾을 수 없습니다."

            print(f"📌 제목: {title}")
            print(f"👤 작가: {author}")
            print(f"📖 내용: {summary}")
            print("-" * 50)

            driver.close()
            driver.switch_to.window(driver.window_handles[0])

        except Exception as e:
            continue

except Exception as e:
    print(f"❌ 에러 발생: {e}")
finally:
    driver.quit()
```
