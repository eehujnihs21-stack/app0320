## **과제 2. Tom이 Jerry를 쫒아가는 행위를 method(함수)로 만들어 보자(10번 following)**

### **class User:**
> **def __init __ (self):**
>   - ##### Initialize attributes
>   - **HAS (Attiributes)**

> **def following_mode():**
>   - ##### change attributes (following, follower) for chasing
>   - **Does (Method, Change the attributes)**

<img src= 'https://pbs.twimg.com/profile_images/810136393/tom-jerry-pic_400x400.jpg' width=300>

```python
class User:
    def __init__(self, name):
        self.name = name
        self.followers = 0
        self.following = 0

    def follow(self, target_user):
        self.following += 1
        target_user.followers += 1
        print(f"※ {self.name}이(가) {target_user.name}을(를) 쫓아갑니다!\n(현재 {self.name}의 추격 횟수: {self.following}회)")

tom = User("Tom")
jerry = User("Jerry")

print("추격전 시작!\n")

for i in range(10):
    tom.follow(jerry)

print("\n--- 최종 결과 ---")
print(f"{tom.name} 팔로잉(추격 중): {tom.following}회")
print(f"{jerry.name} 팔로워(도망 중): {jerry.followers}회")
```
